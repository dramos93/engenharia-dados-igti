# Exercício com os dados do ENEM 2019

```python
url = "http://download.inep.gov.br/microdados/microdados_enem_2019.zip"
```

```python
import pandas as pd
import numpy as np
import zipfile
import requests
from io import BytesIO
import os
```

```python
# Criar um diretório para armazenar os dados do ENEM
os.makedirs('exenem', exist_ok=True)

# Fazer a requisição do arquivo .zip
filebytes = BytesIO(
    requests.get(url, stream=True).content
)

# Extrair o conteúdo do zipfile

myzip = zipfile.ZipFile(filebytes)
myzip.extractall('exenem')
```
