# Transformações

- Co_IES - instituição de ensino
- Co_CATEGAD - categoria administrativa
- CO_GRUPO - grupo do curso
- Co_MODALIDADE - presencial ou ead
- CO_UF_CURSO
- CO_REGIAO_CURSO
- NU_IDADE
- TP_SEXO
- NT_GER
- NT_FG
- NT_CE

```python
enade['DESC_PUBLICA'] = ""
enade.loc[
    enade.CO_CATEGAD.isin([118,120,121,10005,10006,10007,10008,10009,17634]),
    "DESC_PUBLICA"    
] = "Privado"
```

```python
enade['DESC_PUBLICA'] = ""
enade.loc[
    enade.CO_CATEGAD.isin([118,120,121,10005,10006,10007,10008,10009,17634]),
    "DESC_PUBLICA"    
] = "Publico"
```

```python
enade.DESC_PUBLICA.value_counts() / enade.shape[0]
```

```python
# Transformar a modalidade, atribuindo valor a uma coluna existente.
enade["CO_MODALIDADE"] = enade.CO_MODALIDADE.replace({
    0: "EaD",
    1: "Presencial"
})
```

```python
enade.CO_MODALIDADE.value_counts()
```

```python
enade["CO_REGIAO_CURSO"] = enade.CO_REGIAO_CURSO.replace({
    1: "Norte",
    2: "Nordeste",
    3: "Sul",
    4: "Sul",
    5: "Centro-Oeste"
})
```

```python
enade.CO_REGIAO_CURSO.value_counts()
```

```python
# Transformando Cor Raça

enade["DESC_COR"] = enade.QE_I02.replace({
    "A": "Branca",
    "B": "Preta",
    "C": "Amarela",
    "D": "Parda",
    "E": "Indígena",
    "F": pd.NA,
    " ": pd.NA
})
```

```python
dict(enade.DESC_COR.value_counts())
```
